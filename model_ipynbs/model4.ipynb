```
# imports
from math import factorial
from collections import Counter
from itertools import combinations, islice
from scipy import stats
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
from sklearn import preprocessing
warnings.filterwarnings('ignore')
```

```
filename = '/content/drive/Shared drives/ML Project/data/final_train_data.csv'
train_data = pd.read_csv(filename)
```

```

object = dict()
features = ['pid']
for i in range(len(features)) :
    c = features[i]
    object[i] = preprocessing.LabelEncoder()
    train_data[c] = object[i].fit_transform(train_data[c].astype(str))
    
```

```
train_data = train_data.dropna()
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.metrics import average_precision_score, precision_score, accuracy_score

X = train_data[train_data.columns[:-1]].to_numpy()
y = train_data[train_data.columns[-1]].to_numpy()

clf=RandomForestClassifier(n_estimators=700, n_jobs=-1)


kf = KFold(n_splits=4,random_state=42,shuffle=True)
totalAccuracy=[]

for train_index, test_index in kf.split(X):

    X_train   = X[train_index]
    y_train = y[train_index]

    X_test    = X[test_index]
    y_test  = y[test_index]
    
    clf.fit(X_train,y_train)
    predictions = clf.predict(X_test)

    accuracy = accuracy_score(y_test,predictions)
    totalAccuracy.append(accuracy)
    print(accuracy)
    
    
average_accuracy = np.mean(totalAccuracy)
print ("Random Forest Accuracy:")
print (average_accuracy)
```
