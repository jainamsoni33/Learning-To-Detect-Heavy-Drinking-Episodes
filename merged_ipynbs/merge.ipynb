```

df = df.sort_values(by=['time'])
df1 = df1.sort_values(by=['timestamp'])
```

```
index_list = df1.index
index = 0
last_flag = False
df['TAC_Reading'] = 0.0

for row in df.iterrows():

  if index == len(index_list) - 1:
    last_flag = True
    df['TAC_Reading'][row[0]] = df1['TAC_Reading'][index_list[index]]
    continue

  if not last_flag and row[1][0] >= df1['timestamp'][index_list[index]] and row[1][0] < df1['timestamp'][index_list[index+1]]:
    df['TAC_Reading'][row[0]] = df1['TAC_Reading'][index_list[index]]
  else:
    index += 1
    df['TAC_Reading'][row[0]] = df1['TAC_Reading'][index_list[index]]
```
